# 1. The Geometry of Linear Equations

* Linear Algebra is used to solve a system of linear equations
* Simplest case - n equations with n unknowns, and the task is to find these n unknowns.  

# Problem in 2 Dimensions
<br>
<center>2 Equations, 2 Unknowns</center>
$$2x - y = 0$$
$$-x + 2y = 3$$


* Row Solution
    
<center>Equation 1:</center>
$$2x - y = 0$$
$$\therefore y = 2x$$
<br>
<center>Equation 2:</center>
$$-x + 2y = 3$$
$$\therefore y = 0.5x + 1.5$$
<br>
<center>Solving both:</center>
$$\therefore y^* = 2x^* = 0.5x^* + 1.5 $$
$$\therefore x^* = 1, y^* = 2 $$


```python
from numpy import linspace as ls
x = ls(-3, 3, 100)
y1 = 2 * x
y2 = 0.5 * x + 1.5

# Solution
x_ = x[y1 == y2]
y_ = y1[x == x_]
print('Solution:', x_, y_)

from matplotlib.pyplot import * 
from matplotlib import rcParams 
rcParams['lines.markersize'] = 3
style.use('ggplot')
scatter(y1, x, label = 'Equation 1: 2x - y = 0', marker = ',')
scatter(y2, x, label = 'Equation 2: -x + 2y = 3', marker = ',')
scatter(y_, x_, label = 'Solution', marker = 'x')
legend()
show()
```

* Column Picture:
$$x\begin{bmatrix} 2 \\ -1\end{bmatrix} +  y\begin{bmatrix} -1 \\ 2  \end{bmatrix} = \begin{bmatrix} 0 \\ 3\\  \end{bmatrix} $$
<br>
$$ or, ax + by = c $$
<br>
<center>where ax + by is a linear combination known vectors a, b.<center>
<center>The solution are scalars x, y that make ax + by = c.<center>

```python
from numpy import array
from matplotlib.pyplot import * 
a = array([2, -1])
b = array([-1, 2])
c = array([0, 3])

figure(figsize=(12,9))
ylim([-5,5])
xlim([-5,5])

arrow(0, 0, a[0], a[1], width=0.05, color = 'red')
annotate('a', xy=(0, 0), xytext=(a[0]+0.1, a[1]+0.1), arrowprops=dict(arrowstyle="->"))
arrow(0, 0, b[0], b[1], width=0.05, color = 'red')
annotate('b', xy=(0, 0), xytext=(b[0]+0.1, b[1]+0.1), arrowprops=dict(arrowstyle="->"))
arrow(0, 0, c[0], c[1], width=0.05, color = 'green')
annotate('c', xy=(0, 0), xytext=(c[0]+0.1, c[1]+0.1), arrowprops=dict(arrowstyle="->"))

N = 5
x = y = ls(-2, 2, N)
xv, yv = np.meshgrid(x, y, sparse=False, indexing='ij')
for i in range(N):
    for j in range(N):
        lc = a * xv[i,j] + b * yv[i,j]
        arrow(0, 0, lc[0], lc[1], width=0.01, color = 'black')
        annotate(f'x={xv[i,j]},y={yv[i,j]}', xy=(0, 0), xytext=(lc[0]+0.2, lc[1]+0.2), arrowprops=dict(arrowstyle="->"))
        
        if (lc==c).all():
            print(f'solution: x={xv[i,j]}, y={yv[i,j]}')
show()
```

* "The Matrix Picture"
$$\begin{bmatrix} 2 & -1 \\ -1 & 2 \\  \end{bmatrix} \begin{bmatrix} x \\ y\\  \end{bmatrix} = \begin{bmatrix} 0 \\ 3\\  \end{bmatrix} $$
<br>
$$ or, Ax = b $$
<center>where A is fixed coefficient matrix and x is the choice vector. <center>
<br>
$$ \therefore x = A^{-1}b $$



```python
from numpy import array, dot
from numpy.linalg import inv
A = array([[2, -1],
          [-1, 2]])
b = array([[0],
           [3]])
x = dot(inv(A), b)
print(x)
```

# Problem in 3 Dimensions
<br>
<center>3 equations and 3 unknowns</center>
$$2x - y = 0$$
$$-x + 2y - z = -1$$
$$-3y + 4z = 4$$


* Row Solution
    
<center>Equation 1:</center>
$$2x - y = 0$$
$$\therefore y = 2x$$
<br>
<center>Equation 2:</center>
$$-x + 2y - z = -1$$
$$\therefore -x + 2x - z = -1$$
$$\therefore z=-x + 2x + 1 $$
$$\therefore z=x + 1 $$
<br>
<center>Equation 3:</center>
$$-3y + 4z = 4$$
$$\therefore -6x + 4x + 4 = 4$$
<br>

$$\therefore x^* = y^* = 0$$
$$\therefore z^* = 1$$

Each condition is a 2D plane in 3D space. Any two conditions will intersect to give a 2D line in 3D. All three conditions will intersect at a single point, the solution to the problem. However, beyond 2D and 3D space it is hard to visualize the row picture. 


* Column Solution

$$x\begin{bmatrix} 2 \\ -1 \\ 0\end{bmatrix} +  
y\begin{bmatrix} -1 \\ 2 \\ -3 \end{bmatrix} +
z\begin{bmatrix} 0 \\ -1 \\ 4 \end{bmatrix} = 
\begin{bmatrix} 0 \\ 1 \\  4 \end{bmatrix} $$
<br>
$$ or, ax + by + cz = d $$
<br>
where ax + by + cz is a linear combination known vectors a, b and c.
The solution, if it exists, are scalars x, y that make ax + by + cz = d.
    
* "The Matrix Picture"

$$
\begin{bmatrix} 2 & -1 & 0 \\ -1 & 2 & -1 \\0 & 3 & 4 \end{bmatrix} \begin{bmatrix} x \\ y \\ z \\ \end{bmatrix} 
= \begin{bmatrix} 0 \\ -1 \\ 4 \end{bmatrix} $$
<br>
$$ or, Ax = b $$
<center>where A is fixed coefficient matrix and x is the choice vector. <center>
<br>
$$ \therefore x^* = A^{-1}b $$

# Existance of Solution

It is not always that there exists a single x* that satisfies Ax = b. The existance of a solution depends on both A and b. Even if there is a solution it might not be unique.

If Ax is to be able to represent all b, then A matrix has to be: 
* square (necessary)
* nonsingular i.e inverse exists (sufficient)

# Problem in 9 dimensions

* Row Picture - is impossible to visualize. 
* Column picture - is possible to visualize. 
* Matrix picture - is possible to visualize

```python
import numpy as np
A = np.random.normal(0, 1, (9, 9))
b = np.random.normal(0, 1, (9, 1))
x = dot(inv(A), b)
print(A.shape)
```
