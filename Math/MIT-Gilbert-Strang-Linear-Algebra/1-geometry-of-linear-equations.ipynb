# 1. The Geometry of Linear Equations

* Linear Algebra is used to solve a system of linear equations
* Simplest case - n equations with n unknowns, and the task is to find these n unknowns.  

# Problem in 2 Dimensions
<br>
<center>Find x and y, which satisfy: </center>
$$2x - y = 0$$
$$-x + 2y = 3$$


* The Row Picture - 
    
<center>Equation 1:</center>
$$2x - y = 0$$
$$\therefore y = 2x$$
<br>
<center>Equation 2:</center>
$$-x + 2y = 3$$
$$\therefore y = 0.5x + 1.5$$
<br>
<center>Solving both:</center>
$$\therefore y^* = 2x^* = 0.5x^* + 1.5 $$
$$\therefore x^* = 1, y^* = 2 $$


```python
from numpy import arange, meshgrid
X = Y = arange(-15, 15, 0.25) # range of x, y values to test
X, Y = meshgrid(X, Y) # cartesian product
EQ1 = (2 * X - Y == 0) # condition 1
EQ2 = (-X + 2 * Y == 3) # condition 2
print(f'Solution is at x={X[(EQ1*EQ2)].item()}, y={Y[(EQ1*EQ2)].item()}') # x, y that satisfy both conditions

from matplotlib.pyplot import *
figure(figsize=(10,10))
scatter(Y[EQ1], X[EQ1], label = '2x - y = 0', marker = '.') # plot conditions
scatter(Y[EQ2], X[EQ2], label = '-x + 2y = 3', marker = '.')
scatter(Y[(EQ1*EQ2)], X[(EQ1*EQ2)], label = 'Solution', marker = 'x', color = 'black') # mark solution
xlabel('x'); ylabel('y')
title('Row Picture', fontsize=15)
legend(prop={'size': 15})
show()
```

* Column Picture:
$$x\begin{bmatrix} 2 \\ -1\end{bmatrix} +  y\begin{bmatrix} -1 \\ 2  \end{bmatrix} = \begin{bmatrix} 0 \\ 3\\  \end{bmatrix} $$
<br>
$$ or, ax + by = c $$
<br>
<center>where ax + by is a linear combination known vectors a, b.<center>
<center>The solution are scalars x, y that make ax + by = c.<center>

```python
from numpy import array, c_
a = array([2, -1])
b = array([-1, 2])
c = array([0, 3])

O = [0], [0] # origin point
LCs = array([*O])
for x in arange(-3, 3, 0.25): # iterate over scaler x
    for y in arange(-3, 3, 0.25): # iterate over scaler x
        lc = a * x + b * y # calculate linear combination
        LCs = c_[LCs, lc] # store linear combination
        if (lc == c).all(): # test condition
            print(f'Solution: x={x}, y={y}') # solution 

figure(figsize=(10,10))
O = [0], [0] # origin point
V = LCs.T
quiver(*O, V[:,0], V[:,1], color=['black'], scale=21, width=0.002, edgecolors='k', label = 'Linear Combinations')
V = c_[a, b, c].T
quiver(*O, V[:,0], V[:,1], color=['r','b','g'], scale=21)
title('Column Picture', fontsize=15)
legend(prop={'size': 15})
show()
```

* "The Matrix Picture"
$$\begin{bmatrix} 2 & -1 \\ -1 & 2 \\  \end{bmatrix} \begin{bmatrix} x \\ y\\  \end{bmatrix} = \begin{bmatrix} 0 \\ 3\\  \end{bmatrix} $$
<br>
$$ or, Ax = b $$
<center>where A is fixed coefficient matrix and x is the choice vector. <center>
<br>
$$ \therefore x = A^{-1}b $$



```python
from numpy import array, dot
from numpy.linalg import inv
A = array([[2, -1],
          [-1, 2]])
b = array([[0],
           [3]])
x = dot(inv(A), b)
print(f'Solution vector is: \n{x}')
```

# Problem in 3 Dimensions
<br>
<center>3 equations and 3 unknowns</center>
$$2x - y = 0$$
$$-x + 2y - z = -1$$
$$-3y + 4z = 4$$


* Row Solution
    
<center>Equation 1:</center>
$$2x - y = 0$$
$$\therefore y = 2x$$
<br>
<center>Equation 2:</center>
$$-x + 2y - z = -1$$
$$\therefore -x + 2x - z = -1$$
$$\therefore z=-x + 2x + 1 $$
$$\therefore z=x + 1 $$
<br>
<center>Equation 3:</center>
$$-3y + 4z = 4$$
$$\therefore -6x + 4x + 4 = 4$$
<br>

$$\therefore x^* = y^* = 0$$
$$\therefore z^* = 1$$

Each condition is a 2D plane in 3D space. Any two conditions will intersect to give a 2D line in 3D. All three conditions will intersect at a single point, the solution to the problem. However, beyond 2D and 3D space it is hard to visualize the row picture. 


* Column Solution

$$x\begin{bmatrix} 2 \\ -1 \\ 0\end{bmatrix} +  
y\begin{bmatrix} -1 \\ 2 \\ -3 \end{bmatrix} +
z\begin{bmatrix} 0 \\ -1 \\ 4 \end{bmatrix} = 
\begin{bmatrix} 0 \\ 1 \\  4 \end{bmatrix} $$
<br>
$$ or, ax + by + cz = d $$
<br>
where ax + by + cz is a linear combination known vectors a, b and c.
The solution, if it exists, are scalars x, y that make ax + by + cz = d.
    
* "The Matrix Picture"

$$
\begin{bmatrix} 2 & -1 & 0 \\ -1 & 2 & -1 \\0 & 3 & 4 \end{bmatrix} \begin{bmatrix} x \\ y \\ z \\ \end{bmatrix} 
= \begin{bmatrix} 0 \\ -1 \\ 4 \end{bmatrix} $$
<br>
$$ or, Ax = b $$
<center>where A is fixed coefficient matrix and x is the choice vector. <center>
<br>
$$ \therefore x^* = A^{-1}b $$

* It is not always that there exists a single x* that satisfies Ax = b.
* The existance of a solution depends on both A and b. 
* Even if there is a solution it might not be unique.

# Problem in 9 dimensions

* Row Picture - is impossible to visualize. 
* Column picture - is possible to visualize (somehow). 
* Matrix picture - easier to solve, impossible to visualize.

```python
# Matrix Solution
import numpy as np
A = np.random.normal(0, 1, (9, 9))
b = np.random.normal(0, 1, (9, 1))
x = dot(inv(A), b)
print(x)
```
